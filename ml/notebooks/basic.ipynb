## Let's start the analysis

### Import modules.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

### Set the font size globally
plt.rcParams["font.size"] = 18

from sklearn.linear_model import LinearRegression
from scipy import stats
```

### Set the seed and variance of the generative distribution.

Seed is set for the reproducibility. <br>
Set the parametes of the distribution.

```python
np.random.seed(seed=23)

COEFF = 5
SIGMA = 3
```

### Small training data case.

```python
### The number of samples for {training, test}
tr_sample_size = 5
te_sample_size = 50

### Generate samples for training
x_tr = np.random.rand(tr_sample_size, 1)*10
err = SIGMA*np.random.randn(tr_sample_size, 1)
y_tr = COEFF * x_tr  + err

### Generate samples for test
x_te = np.random.rand(te_sample_size, 1)*5 + 15
err = SIGMA*np.random.randn(te_sample_size, 1)
y_te = COEFF * x_te  + err


### Train a linear regression model
slope, intercept, r_value, p_value, std_err = stats.linregress(x_tr.flatten(), y_tr.flatten())

### Make model predctions
xt = np.linspace(0.0, 20, num=1000).reshape((1000, 1))
yt = xt*slope + intercept

### Compute the Mean Squared Error against test samples
SSE = [ (slope*elem[0] + intercept - elem[1])**2 for elem in zip(x_te,y_te)]
MSE = np.sum(SSE) / len(SSE)

### Plot the results
plt.figure(figsize=(14,10))
plt.plot(x_tr, y_tr, "o", color="b", label="Train")
plt.plot(x_te, y_te, "o", color="r", label="Test")
plt.plot(xt, yt, color="k", label="Fitted line")
plt.title("y = 5x + error")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

### Print the fitted results
print("Fitting results")
print( "   slope : {}".format(round(slope,4)) )
print( "   standard error : {}".format(round(std_err,4)) )

print("\nPrediction results")
print( "   MSE : {}".format(round(MSE,4)) )
```
